```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the data from the provided CSV file
file_path = 'F:/FYP project/FYP(Flood)/Example Datasets/river_water_levels_flooding_train_data.csv'
df = pd.read_csv(file_path)

# Extract relevant columns
X = df[['Daily Rainfall (mm)', 'Hourly Water Level (meters)']]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Choose the number of clusters (e.g., 3)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Plotting the clusters
plt.figure(figsize=(10, 8))
plt.scatter(df['Daily Rainfall (mm)'], df['Hourly Water Level (meters)'], c=df['Cluster'], cmap='viridis', alpha=0.5, edgecolors='w', s=70)
plt.title('K-means Clustering of Daily Rainfall vs Hourly Water Level')
plt.xlabel('Daily Rainfall (mm)')
plt.ylabel('Hourly Water Level (meters)')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the data from the provided CSV file
file_path = 'F:/FYP project/FYP(Flood)/Example Datasets/river_water_levels_flooding_train_data.csv'
df = pd.read_csv(file_path)

# Combine Date and Hour columns to create a datetime column
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Hour'].astype(str) + ':00')

# Extract relevant columns for clustering
X = df[['Daily Rainfall (mm)', 'Hourly Water Level (meters)']]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Choose the number of clusters (e.g., 3)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Create scatter plot with clusters
plt.figure(figsize=(16, 7))
scatter = plt.scatter(df['Datetime'], df['Hourly Water Level (meters)'], c=df['Cluster'], cmap='viridis', alpha=0.7, edgecolors='w', s=70)
plt.colorbar(label='Cluster')
plt.xlabel('Datetime')
plt.ylabel('Hourly Water Level (meters)')
plt.title('Hourly Water Level (meters) vs. Date and Hour with Clustering')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```
