```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the data from the provided CSV file
file_path = 'F:/FYP project/FYP(Flood)/Example Datasets/data3.csv'
df = pd.read_csv(file_path)

# Convert month names to numeric values
month_map = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
    'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12
}
df['Month'] = df['Month'].map(month_map)

# Combine Year, Month, and Day into a single datetime column
df['Date'] = pd.to_datetime(df[['Year', 'Month', 'Day']])

# Filter data for the specified months
df_filtered = df[((df['Year'] == 2021) & (df['Month'].isin([4, 5]))) | 
                 ((df['Year'] == 2022) & (df['Month'].isin([4, 5])))]

# Handle missing data: Fill missing values with 0 (assuming no rainfall)
time_columns = ['0830-0900', '0900-1000', '1000-1100', '1100-1200', '1200-1300', '1300-1400', 
                '1400-1500', '1500-1600', '1600-1700', '1700-1800', '1800-1900', '1900-2000', 
                '2000-2100', '2100-2200', '2200-2300', '2300-2400', '0000-0100', '0100-0200', 
                '0200-0300', '0300-0400', '0400-0500', '0500-0600', '0600-0700', '0700-0800', 
                '0800-0830']
df_filtered[time_columns] = df_filtered[time_columns].fillna(0)

# Ensure that all time columns are numeric
df_filtered[time_columns] = df_filtered[time_columns].apply(pd.to_numeric, errors='coerce').fillna(0)

# Melt the DataFrame to long format for clustering
df_melted = df_filtered.melt(id_vars=['Date'], value_vars=time_columns, var_name='Time', value_name='Rainfall')

# Create a datetime column with only the day and hour for plotting
df_melted['Day_Hour'] = df_melted['Date'].dt.strftime('%d') + ' ' + df_melted['Time']

# Prepare data for clustering (features: rainfall amounts for each time slot)
clustering_data = df_filtered[time_columns]

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=0)  # Adjust n_clusters as needed
df_filtered['Cluster'] = kmeans.fit_predict(clustering_data)

# Merge cluster labels with melted data
df_melted = df_melted.merge(df_filtered[['Date', 'Cluster']], on='Date')

# Plotting the clustered data as a scatter plot
plt.figure(figsize=(14, 8))
for cluster in sorted(df_melted['Cluster'].unique()):
    cluster_data = df_melted[df_melted['Cluster'] == cluster]
    plt.scatter(cluster_data['Day_Hour'], cluster_data['Rainfall'], label=f'Cluster {cluster}', alpha=0.7)

plt.title('Clustered Rainfall Data Over Time (April-May 2021 and 2022)')
plt.xlabel('Day and Hour')
plt.ylabel('Rainfall (mm)')
plt.xticks(rotation=90)
plt.legend(title='Cluster', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load the data
file_path = 'F:/FYP project/FYP(Flood)/Example Datasets/data3.csv'
df = pd.read_csv(file_path)

# Inspect the first few rows and columns
print(df.head())
print(df.info())
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())

# Handle missing values by filling them or removing rows/columns with missing data
df.fillna(0, inplace=True)  # Filling missing values with 0 for rainfall data

# Ensure all time columns are numeric
time_columns = ['0830-0900', '0900-1000', '1000-1100', '1100-1200', '1200-1300', '1300-1400', 
                '1400-1500', '1500-1600', '1600-1700', '1700-1800', '1800-1900', '1900-2000', 
                '2000-2100', '2100-2200', '2200-2300', '2300-2400', '0000-0100', '0100-0200', 
                '0200-0300', '0300-0400', '0400-0500', '0500-0600', '0600-0700', '0700-0800', 
                '0800-0830']
df[time_columns] = df[time_columns].apply(pd.to_numeric, errors='coerce').fillna(0)
```

```python
# Convert month names to numeric values if not already done
month_map = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
    'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12
}
df['Month'] = df['Month'].map(month_map)

# Combine Year, Month, and Day into a single datetime column
df['Date'] = pd.to_datetime(df[['Year', 'Month', 'Day']])
df['Day_of_Week'] = df['Date'].dt.day_name()
df['Hour'] = df['Date'].dt.hour
```
